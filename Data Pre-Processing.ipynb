# Building a Master Dataset of Required Information

```python
# set the start date and end date.
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2020,1,1)
BAC = pd.read_csv('./Dataset/BAC.csv')
C = pd.read_csv('./Dataset/C.csv')
GS = pd.read_csv('./Dataset/GS.csv')
JPM = pd.read_csv('./Dataset/JPM.csv')
MS = pd.read_csv('./Dataset/MS.csv')
WFC = pd.read_csv('./Dataset/WFC.csv')
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)

#Setting the column name levels:
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
bank_stocks.head()
```

```python
bank_stocks.to_csv(r'.\Big6.csv')
```

# Generating the Return Dataframe

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()
returns['Date'] = bank_stocks['BAC']['Date'] 
returns.head(3)
```

```python
returns.to_csv('returns_data.csv', index=False)
```

# Nested Pie Chart Dataset

```python
df=pd.read_csv(".\Big6.csv")
df
```

```python
df=df.drop(0)
df.drop(["Bank Ticker"],axis=1)
df.head()
```

```python
date_Columns = ["BAC","C","GS","JPM","MS","WFC"]
num_columns = ['BAC.1', 'BAC.2', 'BAC.3', 'BAC.4', 'BAC.5','BAC.6','C.1', 'C.2', 'C.3', 'C.4', 'C.5', 'C.6','GS.1','GS.2', 'GS.3', 'GS.4', 'GS.5', 'GS.6','JPM.1', 'JPM.2','JPM.3', 'JPM.4', 'JPM.5', 'JPM.6','MS.1', 'MS.2', 'MS.3','MS.4', 'MS.5', 'MS.6','WFC.1', 'WFC.2', 'WFC.3', 'WFC.4','WFC.5', 'WFC.6']
df[date_Columns] = df[date_Columns].apply(pd.to_datetime, errors='coerce')
df[num_columns] = df[num_columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.describe()
```

```python
banks = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
new_df = pd.DataFrame(columns=['Bank', 'Type', 'Avg'])
```

```python

new_df = pd.DataFrame()  # Create an empty DataFrame to store the averages

for bank in banks:
    # Get the 'High' and 'Close' columns for the current bank
    high_col = bank + '.1'
    low_col = bank + '.2'
    open_col = bank + '.3'
    close_col = bank + '.4'
    volume_col = bank + '.5'
    adjClose_col = bank+'.6'
    
    
    high_avg = df[high_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['High'], 'Avg': [round(high_avg)]})], ignore_index=True)
    
    low_avg = df[low_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['Low'], 'Avg': [round(low_avg)]})], ignore_index=True)
    
    open_avg = df[open_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['Open'], 'Avg': [round(open_avg)]})], ignore_index=True)
    
    close_avg = df[close_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['Close'], 'Avg': [round(close_avg)]})], ignore_index=True)
    
    
    adjclose_avg = df[adjClose_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['Adj Close'], 'Avg': [round(adjclose_avg)]})], ignore_index=True)
    
    volume_avg = df[volume_col].mean()
    new_df = pd.concat([new_df, pd.DataFrame({'Bank': [bank], 'Type': ['Volume'], 'Avg': [round(volume_avg)]})], ignore_index=True)
    
```

```python
new_df
```

```python
new_df.to_csv("Pie Chart data.csv")
```
